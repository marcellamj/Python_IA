```python
'''
LabelEncoder - transformar objetos em números
Criar codificador
fit_transform - vai salvar como número e vice-versa, não precisa definir os números


profissao                 object
mix_credito               object
comportamento_pagamento   object
'''
```

```python
# Passo 2
from sklearn.preprocessing import LabelEncoder

# Criar codificador
codif = LabelEncoder()

# fit_transform - vai salvar como número e vice-versa, não precisa definir os números
df["profissao"]= codif.fit_transform(df["profissao"])
df["mix_credito"]= codif.fit_transform(df["mix_credito"])
df["comportamento_pagamento"]= codif.fit_transform(df["comportamento_pagamento"])

df.info()
```

```python
# Separar dados de treino e de teste
# axis=0 - linha
# axis=1 - coluna

# O que eu quero prever/gabarito para checar acurácia da previsão
y = df['score_credito']

# O que eu vou usar para fazer a previsão - quanto menos informação melhor para o modelo
x = df.drop(['id_cliente','score_credito'],axis=1)
```

```python
# Biblioteca que separa os modelos de treino e teste
from sklearn.model_selection import train_test_split
# Precisa ser na ordem abaixo:
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

```python
# Passo 3 - criar os modelos de IA - Score Ruim, Médio, Bom e ver qual é o melhor:
# Importar biblioteca de Arvore de Decisão
# KNN - nearest neighbors
# Possuem parâmetros para melhoria do modelo

from sklearn.ensemble import RandomForestClassifier
# DecisionTreeClassifier - é outro modelo de árvore, mais simples

from sklearn.neighbors import KNeighborsClassifier

modelo_arvore=RandomForestClassifier()
modelo_knn=KNeighborsClassifier()

modelo_arvore.fit(x_treino,y_treino)
modelo_knn.fit(x_treino,y_treino)
```

```python
# Passo 4 - Escolher o melhor modelo
# Importar biblioteca de Acurácia do modelo
from sklearn.metrics import accuracy_score

# Criar variavel de previsao com os modelos
previsao_arvore=modelo_arvore.predict(x_teste)
previsao_knn=modelo_knn.predict(x_teste.to_numpy())
# Precisa tranformar o df pandas em numpy

# Printar acurácia com o gabarito y_teste
print(accuracy_score(y_teste, previsao_arvore))
print(accuracy_score(y_teste, previsao_knn))
```

**O melhor modelo foi o da árvore de decisão - 83% acurácia**
